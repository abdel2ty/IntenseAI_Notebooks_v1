# **1Ô∏è‚É£ Grouping in Pandas ‚Äî `.groupby()`**

The **`groupby()`** method splits your data into groups based on one or more columns, then you can **aggregate**, **transform**, or **apply custom functions**.

### **Basic Syntax**

```python
df.groupby("column_name")
df.groupby(["col1", "col2"])
```

## **1.1 Group by One Column**

### Example ‚Üí Total cases per state




````python
df.groupby("state")["cases"].sum()
````




## **1.2 Group by Multiple Columns**

**total cases per county inside each state**.

````python
#pd.set_option("display.max_rows", None)
#pd.reset_option("display.max_rows")
````

````python
df.groupby(["state", "county"])["cases"].sum()
````

````python
df.groupby(["state", "county"])["cases"].mean().sort_values(ascending=False)
````


## **1.3 Using Multiple Aggregations**

Use **`.agg()`** to compute several metrics at once:


````python
df.groupby("state").agg(
    {
    "cases": ["sum", "mean", "max"],
    "deaths": ["sum", "mean"]
})
````

## **1.4 Group & Count**

Count the number of rows per group:

````python
df.groupby("state")["county"].value_counts()
````

````python
df.groupby("state")["county"].value_counts()
````


## **1.5 Group & Filter**

Find states with **more than 800 total cases**:


Sometimes, instead of collapsing groups, we want to **add new columns** based on group statistics:


````python
df["state_avg_cases"] = df.groupby("state")["cases"].transform("mean")
df
````


# üîπ **Understanding `how` in `pd.merge()`**

When we merge two **DataFrames** in Pandas using `pd.merge()`, the **`how`** parameter **controls the type of join** ‚Äî just like **SQL joins**.

---

## **üìå Syntax**

```python
pd.merge(left_df, right_df, on="common_column", how="type_of_join")
```

* **`left_df`** ‚Üí First DataFrame
* **`right_df`** ‚Üí Second DataFrame
* **`on`** ‚Üí Column name to merge on
* **`how`** ‚Üí Type of join (controls which rows are included)

---

## **üîπ Types of Joins (`how`)**

| **Type**              | **Keeps Rows From**         | **Keeps Unmatched from Left** | **Keeps Unmatched from Right** | **Result Size**           |
| --------------------- | --------------------------- | ----------------------------- | ------------------------------ | ------------------------- |
| **inner** *(default)* | Only matching rows          | ‚ùå No                          | ‚ùå No                           | Smallest                  |
| **left**              | All rows from **left\_df**  | ‚úÖ Yes                         | ‚ùå No                           | At least as many as left  |
| **right**             | All rows from **right\_df** | ‚ùå No                          | ‚úÖ Yes                          | At least as many as right |
| **outer**             | All rows from **both**      | ‚úÖ Yes                         | ‚úÖ Yes                          | Largest                   |

---

## **üîπ Example DataFrames**

### **Main Data (df)**

| county  | cases |
| ------- | ----- |

| county  | population |

## **1Ô∏è‚É£ Inner Join (Default)**

```python
df_merged = pd.merge(df, df_population, on="county", how="inner")
print(df_merged)
```

**Result:**

| county  | cases | population |
| ------- | ----- | ---------- |
| Autauga | 18961 | 55869      |
| Baldwin | 67496 | 227131     |

‚úÖ Keeps **only counties present in both DataFrames**.

---

## **2Ô∏è‚É£ Left Join**

```python
df_merged = pd.merge(df, df_population, on="county", how="left")
print(df_merged)
```

**Result:**

| county  | cases | population |
| ------- | ----- | ---------- |
| Autauga | 18961 | 55869      |
| Baldwin | 67496 | 227131     |
| Barbour | 7027  | NaN        |
| Bibb    | 7692  | NaN        |

‚úÖ Keeps **all rows from `df` (left)**.
‚ùå If a county is **missing in `df_population`**, we get **NaN**.

---

## **3Ô∏è‚É£ Right Join**

```python
df_merged = pd.merge(df, df_population, on="county", how="right")
print(df_merged)
```

**Result:**

| county  | cases | population |
| ------- | ----- | ---------- |
| Autauga | 18961 | 55869      |
| Baldwin | 67496 | 227131     |
| Blount  | NaN   | 59852      |

‚úÖ Keeps **all rows from `df_population` (right)**.
‚ùå If a county is **missing in `df`**, `cases` will be **NaN**.

---

## **4Ô∏è‚É£ Outer Join**

```python
df_merged = pd.merge(df, df_population, on="county", how="outer")
print(df_merged)
```

**Result:**

| county  | cases | population |
| ------- | ----- | ---------- |
| Autauga | 18961 | 55869      |
| Baldwin | 67496 | 227131     |
| Barbour | 7027  | NaN        |
| Bibb    | 7692  | NaN        |
| Blount  | NaN   | 59852      |

‚úÖ Keeps **all rows from both DataFrames**.
‚ùå Missing values become **NaN**.

---

## **üìå Summary of `how`**

| **how** | **Keeps all from Left** | **Keeps all from Right** | **Keeps only Matches** |
| ------- | ----------------------- | ------------------------ | ---------------------- |
| inner   | ‚ùå                       | ‚ùå                        | ‚úÖ                      |
| left    | ‚úÖ                       | ‚ùå                        | ‚úÖ                      |
| right   | ‚ùå                       | ‚úÖ                        | ‚úÖ                      |
| outer   | ‚úÖ                       | ‚úÖ                        | ‚úÖ                      |

---



## ** 1. Overview**

| **Feature**          | **`pd.concat()`** üü¢                                                                   | **`pd.merge()`** üîµ                                                          |
| -------------------- | -------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------- |
| **Purpose**          | Combines **DataFrames** **on top of each other** (rows) or **side by side** (columns). | Combines DataFrames **based on common columns or indexes** (like SQL joins). |
| **Join Type**        | Simple stacking, **no matching needed** unless specified.                              | Smart combining, **matches rows based on keys**.                             |
| **Default Behavior** | Vertical stacking (adds rows).                                                         | Inner join (keeps matching rows).                                            |
| **Axis**             | You can specify `axis=0` (rows, default) or `axis=1` (columns).                        | Always merges **based on a key**, not axes.                                  |
| **Use Case**         | Adding **new rows** or **new columns** when structure is similar.                      | Combining **related datasets** using a common key.                           |

---

## ** 2. Using `pd.concat()`** ‚Üí **Stacking Data**


### **üîπ Example 1 ‚Äî Vertical Concatenation (Add Rows)**

```python
df_concat = pd.concat([df1, df2], axis=0)
print(df_concat)
```


### **üîπ Example 2 ‚Äî Horizontal Concatenation (Add Columns)**
```python
df_concat_cols = pd.concat([df1, df3], axis=1)
print(df_concat_cols)
```

## **üîπ  Using `pd.merge()` ** ‚Üí **Smart Combining Based on Keys**


### **üîπ Example 3 ‚Äî Merge on a Common Column**

```python
df_merged = pd.merge(df1, df_population, on="county", how="left")
print(df_merged)
```


## **üîπ  When to Use What**

* ‚úÖ **Use `concat()`** when you just want to **append rows** or **stack columns**.
* ‚úÖ **Use `merge()`** when you want to **combine datasets intelligently** based on **common columns** (like SQL joins).
